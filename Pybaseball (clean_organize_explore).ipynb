```python
# Import packages and libraries

from pybaseball import batting_stats_range
from pybaseball import batting_stats
from pathlib import Path
import numpy as np
import pandas as pd
```

```python
# Gathering batting stats from 2015

data = batting_stats(2015)

batting_df = pd.DataFrame(data)
batting_df.head()
```

```python
len(data)
```

```python
# Dropped all columns with NaN values 

batting_df = batting_df.dropna(axis=1)
batting_df
```

```python
batting_df = batting_df.drop(columns= ['IDfg','Season','Team','Age'])
batting_df
```

```python
batting_df.describe()
```

```python
# Adding a 'MVP' column

batting_df["MVP"] = "np.NaT"
batting_df
```

```python
# Changing the values in the 'MVP' column to '0'.

batting_df['MVP'] = 0
batting_df.head()
     
```

```python
# This dataset is for the year 2015. Josh Donaldson won MVP that year - assigning a '1' to his 'MVP' column.

batting_df.at[5, 'MVP']=1
batting_df.head()
```

```python
# Converting batting_df into a CSV and saving to branch/repository 

batting_df.to_csv('batting_df.csv')
```
