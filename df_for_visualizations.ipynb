# Using cleaned_merged_df to create some specific dataframes for visualizations 

```python
# Importing dependacies/libraries 

from Cleaner import *
```

```python
# Reading in cleaned_merged_df.csv as a dataframe

cleaned_merged_df = pd.read_csv('Resources/cleaned_merged_df.csv', index_col=False)
cleaned_merged_df
```

```python
# Creating a dataframe of just the MVPs

mvps = cleaned_merged_df.loc[cleaned_merged_df['MVP'] == 1 ]
mvps
```

```python
# Drop Unnamed: 0 column 

mvps.drop(columns= 'Unnamed: 0')
```

```python
# Isolating a handful of stats that I deem most important/impactful 

mvps_important_stats = mvps[['H','RBI','AVG','OBP','SLG','OPS','ISO']]
mvps_important_stats
```

```python
# Get summmary statistics 

mvps_important_stats.describe()
```

```python
# Repeat this process to create the same dataframe for non-MVP players

# Creating a dataframe of non-MVPs

non_mvps = cleaned_merged_df.loc[cleaned_merged_df['MVP'] == 0 ]
non_mvps
```

```python
# Drop Unnamed: 0 column 

non_mvps.drop(columns= 'Unnamed: 0')
```

```python
# Isolating a handful of stats that I deem most important/impactful 

non_mvps_important_stats = non_mvps[['H','RBI','AVG','OBP','SLG','OPS','ISO']]
non_mvps_important_stats
```

```python
# Get summmary statistics 

non_mvps_important_stats.describe()
```

```python
# Saving each new dataframe to be used in Tableau 

mvps_important_stats.to_csv('Resources/mvps_stats.csv')
non_mvps_important_stats.to_csv('Resources/non_mvps_stats.csv')
```
