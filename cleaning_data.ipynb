```python
people_df['Name'] = people_df['nameFirst'] + ' ' + people_df['nameLast']
people_df.head()
```

```python
people_df = people_df[['playerID', 'Name']]
people_df.head()
```

```python
df = pd.merge(mvp_df, people_df, on='playerID', how='left')
df
```

```python
mvp_df = df.loc[df['yearID'] >= 2008]
mvp_df = mvp_df.loc[mvp_df['yearID'] < 2021]
mvp_df
```

```python
mvp_df = mvp_df[mvp_df['awardID'] == 'Most Valuable Player']
mvp_df
```

```python
mvp_df = mvp_df.drop(columns=['tie', 'notes'])
mvp_df.head()
```

```python
mvp_df_2008 = mvp_df.loc[mvp_df['yearID'] == 2008]
mvp_df_2008
```

```python
cleaned_df = pd.merge(batting_df_2008, mvp_df_2008, on='Name', how='left')
cleaned_df
```
