```python
from Cleaner import *
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading in cleaned_merged_df.csv as a dataframe

cleaned_merged_df = pd.read_csv('Resources/cleaned_merged_df.csv', index_col=False)
cleaned_merged_df
```

```python
# Creating a dataframe of just the MVPs

mvps = cleaned_merged_df.loc[cleaned_merged_df['MVP'] == 1 ]
mvps
```

```python
# Drop Unnamed: 0 column 

mvps = mvps.drop(columns= 'Unnamed: 0')
```

```python
# Isolating a handful of stats that I deem most important/impactful 

mvps_important_stats = mvps[['H','RBI','AVG','OBP','SLG','OPS','ISO', 'XBH+']]
mvps_important_stats
```

```python
# Repeat this process to create the same dataframe for non-MVP players

# Creating a dataframe of non-MVPs

non_mvps = cleaned_merged_df.loc[cleaned_merged_df['MVP'] == 0 ]
non_mvps
```

```python

# Drop Unnamed: 0 column 

non_mvps = non_mvps.drop(columns= 'Unnamed: 0')
```

```python
# Isolating a handful of stats that I deem most important/impactful 

non_mvps_important_stats = non_mvps[['H','RBI','AVG','OBP','SLG','OPS','ISO','XBH+']]
non_mvps_important_stats
```

```python
# Saving each new dataframe to be used in Tableau 

mvps_important_stats.to_csv('Resources/mvps_stats.csv')
non_mvps_important_stats.to_csv('Resources/non_mvps_stats.csv')
```

```python
mvps_hits = mvps[['H', '1B', '2B', '3B', 'HR', 'R', 'RBI','XBH+']].mean()
mvps_hits
```

```python
x_axis = ['H', '1B', '2B', '3B', 'HR', 'R', 'RBI']
y_axis1 = [mvps['H'].mean(),mvps['1B'].mean(),mvps['2B'].mean(),mvps['3B'].mean(),mvps['HR'].mean(),mvps['R'].mean(),mvps['RBI'].mean()]
y_axis2 = [non_mvps['H'].mean(),non_mvps['1B'].mean(),non_mvps['2B'].mean(),non_mvps['3B'].mean(),non_mvps['HR'].mean(),non_mvps['R'].mean(),non_mvps['RBI'].mean()]
```

```python
comparing_basic_stats = plt.plot(x_axis, y_axis1, y_axis2, 'o')
label = 'MVPs', 'non-MVPs'
plt.title('MVPs VS non-MVPs: Classic Counting Stats')
plt.grid('True')
plt.legend(label)
```
